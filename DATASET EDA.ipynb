```python
sns.boxplot(x='Brand', y='Price', data=df)
plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Box Plot of Price for each Brand')
plt.show()
```

### Observation:- 
<div class="alert alert-block alert-success">

 1.The box plot will show the distribution of car prices for each brand, providing a visual comparison of the spread of prices among different car manufacturers. Some brands may have a larger price range, indicating more variability in their car prices, while others might have a narrower range, suggesting more consistent pricing.

 2.The horizontal line inside the box represents the median price of each brand. It gives an idea of the central tendency of car prices for a particular manufacturer. Comparing the medians between brands can provide insights into the relative pricing levels among different car makers.

 3.The shape of the box and whiskers on the plot can give you an understanding of the price distribution for each brand. If the box is longer, it indicates that most car models for that brand fall within a narrower price range. Conversely, a wider box suggests a broader price distribution.
</div>



```python
sns.countplot(x='Color', data=df)
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Bar Plot of Color')
plt.show()
```

### Observation:<br>

<div class="alert alert-block alert-success">
 1. The bar plot will show the frequency of each car color in the dataset. You will notice a bar for each color, and the height of each bar represents the number of cars with that specific color.

2.White as the Most Common Color: White will have the tallest bar, indicating that it is the most frequently occurring color in the dataset. This suggests that a significant number of cars in the dataset are white.

3.Yellow as the Least Common Color: Yellow will have the shortest bar, indicating that it is the least frequently occurring color in the dataset. This suggests that yellow cars are relatively rare compared to other colors.

4.In some cases, certain colors might be more popular during specific seasons. For instance, brighter colors like yellow might be more common in the summer months, while darker colors like black or blue might be preferred during the winter. The bar plot can help identify such seasonal trends.
</div>

```python
manufacturing_years = df['Year'].value_counts()
plt.pie(manufacturing_years,labels= manufacturing_years.index,autopct="%1.1f%%")
plt.title("Manufacturing years")
plt.show()
```

### Observation:- 
<div class="alert alert-block alert-success">1. The pie chart shows that the dataset has an approximately even distribution of vehicles across different manufacturing years.<br><br>
    2. The similar percentages for manufacturing years 2017, 2019, and 2018 (around 21%) indicate a consistent demand for vehicles during these years. This might be due to popular models or favorable market conditions during that period. <br><br>
    3.The lower percentages for manufacturing years 2015 and 2016 (8.5% and 14.1% respectively) suggest that older models are less represented in the dataset. This could be due to natural aging of vehicles or a decline in demand for those specific models.
</div>

```python
corr_matrix = df[['Year','Mileage','Price']].corr()
sns.heatmap(corr_matrix,annot=True,cmap='PuBu_r')
plt.title("Correlation Matrix")
plt.show()
```

### Observation:-

<div class="alert alert-block alert-success">1. A positive correlation would suggest that newer vehicles tend to have higher prices, while a negative correlation would imply that older vehicles have higher prices due to rarity or vintage value.<br><br>
 2. A negative correlation between mileage and price would indicate that higher mileage vehicles tend to have lower prices, reflecting the impact of wear and tear on vehicle value.<br><br>
3. A positive correlation between these two variables would suggest that older vehicles tend to have higher mileage, which is a common trend due to natural usage over time.</div>
